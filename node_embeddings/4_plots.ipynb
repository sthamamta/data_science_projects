## TSNE plots

```python
import pandas as pd
import numpy as np
import plot
```

```python
nodes = pd.read_csv('data/cora.content', sep='\t', header=None)
nodes = nodes.rename(columns={0:'id', 1434:'class'})
y = nodes[['id','class']]
y = y.sort_values(by=['id'])
y = np.ravel(np.array(y[['class']]))
y
```

## [128,80,50,..,..]

```python
emb_df = pd.read_csv('embedding/[128, 80, 50, 0.25, 0.25].txt', sep=' ', skiprows=[0], header=None)
emb_df = emb_df.sort_values(by=[0])
emb_df = emb_df.drop(emb_df.columns[0], axis=1)
emb_df
plot.draw_plot(emb_df, y, 'images/[128, 80, 50, 0.25, 0.25].png')
```
