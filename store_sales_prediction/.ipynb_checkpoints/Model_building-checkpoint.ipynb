# Handling categorical columns

```python
def print_df_info(df):
    # Number of rows
    num_rows = len(df)

    categorical_columns = df.select_dtypes(include=['object']).columns
    numerical_columns = df.select_dtypes(exclude=['object']).columns

    duplicate_rows = df[df.duplicated()]
    num_duplicate_rows = len(duplicate_rows)

    null_value_count = df.isnull().sum().sum()

    total_memory_size = np.round(df.memory_usage(deep=True).sum()/ (1024**3) )

    print(f"Number of rows: {num_rows}")
    print(f"Number of categorical columns: {len(categorical_columns)}")
    print(f"Number of numerical columns: {len(numerical_columns)}")
    print(f"Duplicate rows count: {num_duplicate_rows}")
    print(f"Null value count: {null_value_count}")
    print(f"Total memory size for the DataFrame: {total_memory_size} GB")



def count_unique_values(df):
    unique_counts = {}
    
    for column in df.columns:
        if df[column].dtype == 'object':
            unique_values = df[column].unique()
            num_unique = len(unique_values)
            unique_counts[column] = num_unique
            
    return unique_counts
```

```python
print_df_info(df)
```

```python
unique_counts = count_unique_values(df)
print(unique_counts)
```
